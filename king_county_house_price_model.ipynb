```python
## Drop the "id" column from the dataset.
df = df.drop("id", axis = 1)
```

```python
## Convert the "date" data into "datetime" data type.
df["date"] = pd.to_datetime(df["date"])
```

```python
## Extract "year" data.
df["year"] = df["date"].apply(lambda date: date.year)
df["month"] = df["date"].apply(lambda date: date.month)
```

```python
## Investigate if month is correlated with price.
plt.figure(figsize = (12, 8))
sns.boxplot(x = "month", y = "price", data = df)
```

```python
## Investigate if month is correlated with price.
df.groupby("month").mean()["price"].plot()
```

```python
## Investigate if year is correlated with price.
df.groupby("year").mean()["price"].plot()
```

```python
## Original date column is not useful for us now.
df = df.drop("date", axis = 1)
```

```python
## Zipcode information is not useful for us.
df = df.drop("zipcode", axis = 1)
```

```python
x = df.drop("price", axis = 1).values
y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(19, activation = "relu"))
model.add(Dense(19, activation = "relu"))
model.add(Dense(19, activation = "relu"))
model.add(Dense(19, activation = "relu"))

model.add(Dense(1))
         
model.compile(optimizer = "adam", loss = "mse")
```

```python
model.fit(x = x_train, y = y_train,
          validation_data = (x_test, y_test),
          batch_size = 128,
          epochs = 800
         )
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```

```python
explained_variance_score(y_test, predictions)
```

```python
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, "r")
```
