```python
## Percentage of missing data.
100 * df.isnull().sum() / len(df)
```

```python
## These features have the most missing data. They will be investigated.
feat_info("emp_title")
feat_info("emp_length")
feat_info("mort_acc")
```

```python
## Investigation of the "emp_title" feature.
len(df["emp_title"].unique()) ## Too many different job titles.
df["emp_title"].value_counts() ## There are many special job titles as well.
## This feature is realistically not useful. It will be removed.
df = df.drop("emp_title", axis = 1)
```

```python
## Investigation of the "emp_length" feature.
sorted(df["emp_length"].dropna().unique())
emp_length_order = ['< 1 year',
 '1 year',
 '2 years',
 '3 years',
 '4 years',
 '5 years',
 '6 years',
 '7 years',
 '8 years',
 '9 years',
 '10+ years']
plt.figure(figsize = (12, 8))
sns.countplot(x = "emp_length", data = df, order = emp_length_order, hue = "loan_status")
## To understand if this feature is informative, we need the percentage of charge off per category.
emp_co = df[df["loan_status"] == "Charged Off"].groupby("emp_length").count()["loan_status"]
emp_fp = df[df["loan_status"] == "Fully Paid"].groupby("emp_length").count()["loan_status"]
emp_len = emp_co / (emp_fp + emp_co)
plt.figure(figsize = (12, 8))
emp_len.plot(kind = "bar")
## This feature turns out to be not useful. It will be dropped.
df = df.drop("emp_length", axis = 1)
```

```python
## The remaining features with missing data will be dealt with now.
df.isnull().sum()
```

```python
## This feature is not useful. It will be dropped.
feat_info("title")
df = df.drop("title", axis = 1)
```

```python
## This feature is missing a lot of data. The corellations will be invesitaged to try and fill in the data.
feat_info("mort_acc")
df["mort_acc"].value_counts()
df.corr()["mort_acc"].sort_values() ## "total_acc" feature positively correlates with "mort_acc" feature.
total_acc_avg = df.groupby("total_acc").mean()["mort_acc"]

def fill_mort_acc(total_acc, mort_acc):
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc

df["mort_acc"] = df.apply(lambda x : fill_mort_acc(x["total_acc"], x["mort_acc"]), axis = 1)
```

```python
## "revol_util" and "pub_rec_bankruptcies" features have a few missing values. The mssing rows will be dropped, features will be kept.
df.isnull().sum()
```

```python
## Drop the missing rows.
df = df.dropna()
```

```python
## The dataset has no missing values anymore.
df.isnull().sum()
```

```python
feat_info("term")
df["term"].value_counts()
df["term"] = df["term"].apply(lambda term: int(term[:3]))
```

```python
## Drop the "grade" feature.
df = df.drop("grade", axis = 1)
```

```python
## Convert "sub_grade" feature into dummy variables.
dummies = pd.get_dummies(df["sub_grade"], drop_first = True)

df = pd.concat([df.drop("sub_grade", axis = 1), dummies], axis = 1)
```

```python
## Convert "verification_status", "application_type", "initial_list_status", "purpose" features into dummy variables.
dummies = pd.get_dummies(df[["verification_status", "application_type", "initial_list_status", "purpose"]], drop_first = True)

df = pd.concat([df.drop(["verification_status", "application_type", "initial_list_status", "purpose"], axis = 1), dummies], axis = 1)
```

```python
## Since there are very few "NONE" and "ANY", put them into "OTHER".
df["home_ownership"].value_counts()
df["home_ownership"] = df["home_ownership"].replace(["NONE", "ANY"], "OTHER")

## Convert "home_ownership" feature into dummy variables.
dummies = pd.get_dummies(df["home_ownership"], drop_first = True)

df = pd.concat([df.drop("home_ownership", axis = 1), dummies], axis = 1)
```

```python
## Extract the xip codes from the "address" feature and turn them into dummy variables. 
df["zip_code"] = df["address"].apply(lambda address: address[-5 :])
df["zip_code"].value_counts()

## Convert "zip_code" feature into dummy variables.
dummies = pd.get_dummies(df["zip_code"], drop_first = True)

df = pd.concat([df.drop("zip_code", axis = 1), dummies], axis = 1)

## Drop the "address" feature.
df = df.drop("address", axis = 1)
```

```python
## Drop the "issue_d" feature.
df = df.drop("issue_d", axis = 1)
```

```python
## Extract the year values from "earliest_cr_line"
feat_info("earliest_cr_line")
df["earliest_cr_line"] = df["earliest_cr_line"].apply(lambda date: int(date[-4 :]))
```

```python
from sklearn.model_selection import train_test_split
```

```python
## Drop the "loan_status" column since we will use "loan_repaid" column instead.
df = df.drop("loan_status", axis = 1)
```

```python
X = df.drop("loan_repaid", axis = 1).values
y = df["loan_repaid"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(78, activation = "relu"))
model.add(Dropout(0.2))

model.add(Dense(39, activation = "relu"))
model.add(Dropout(0.2))

model.add(Dense(19, activation = "relu"))
model.add(Dropout(0.2))

model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x = X_train, y = y_train, epochs = 25, batch_size = 256, validation_data = (X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
losses
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predictions = model.predict_classes(X_test)
print(classification_report(y_test, predictions))
```

```python
confusion_matrix(y_test, predictions)
```
